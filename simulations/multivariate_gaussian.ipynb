```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="dark")
```

```python
def generate_multivariate_plots(
        num_samples=1000,
        mean = [0,0],
        covariance = [(2,0.4),(0.4,0.2)]
):
    rng = np.random.RandomState(0)
    x,y = rng.multivariate_normal(mean,covariance,num_samples).T

    f, ax = plt.subplots(figsize=(6,6))
    sns.scatterplot(x=x,y=y,s=5,color="0.15")
    sns.histplot(x=x,y=y,bins=50,pthresh=0.1,cmap="mako")
    sns.kdeplot(x=x,y=y,levels=5,color='r',linewidths=1)
```

```python
generate_multivariate_plots()
```
