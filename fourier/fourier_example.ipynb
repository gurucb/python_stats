```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Simple code to generate Sine and Cosine. 
1. Using linespace start and stop, create "num" number of points.
2. Default values: Start=0, stop 2*pi. LineSpace distance (radians). Num=100, between start and stop created 100 data points all equi distant.
3. Applied Sine and Cosine functions to input NDArray
4. Using both Sine and Cosine NDArrays plotting graphs

```python
def generate_trig_plots(start=0,stop=2*np.pi,num=100):
    x = np.linspace(start=start,stop=stop,num=num)
    y_sin = np.sin(x)
    y_cos = np.cos(x)

    plt.figure(figsize=(20,6))
    plt.plot(x,y_sin,label="Sine",color="blue")
    plt.plot(x,y_cos,label="Cosine",color="red")

    plt.title("Sine & Cosine Curves")
    plt.xlabel("x (radians)")
    plt.ylabel('y')
    plt.legend()

    plt.show()
```

While both plots are for Sine and Cosine as Number of points in line space are only 10, graph is coarse.

```python
generate_trig_plots(0,8*np.pi,num=10)
```

Plots more smooth as number of points are 100 between 0 and 2 * pi.

```python
generate_trig_plots(0,2*np.pi,num=100)
```

Now plot multiple cycles and notice phase difference between Sine and Cosine plots

```python
generate_trig_plots(
    start=0,
    stop=50*np.pi,
    num=1000
    )
```

Parameters
1. Start: Initial value
2. Stop: Final Value. Between INtial and final values, 'num' number of points created.
3. Freq: (1 .. n) * PI, bascially controlling how many times sine / consine functions oscilate per time period.
4. ampl: Not yet implemented but should be same size of frequency. This is to increase / decrease amplitude of each wave.

Using all these parameters, Sine, Cosine data points are generated, summed up and graphs are plotted.


```python
def generate_complex_trig_plots(start=0,stop=2*np.pi,freq=[1,2,3,4,5],ampl=[1,0,1.0,0,1.0],num=100):
    x = [ np.linspace(start=0,stop= (f * 2  * np.pi),num=100) for f in freq ]
    sin_x = [ np.sin(xi) for xi in x]
    cos_x = [ np.cos(xi) for xi in x]
    fig, axes = plt.subplots(2,1, figsize = (20,10) )
    sign = 1
    for i in range(len(sin_x)):
        axes[0].plot(x[0],(sign/np.pow((i+1),1)) * sin_x[i],color="blue")
        # axes[0].plot(x[0],cos_x[i],color="red")
        # plt.title("Sine & Cosine Curves")

    comb_sin = np.zeros(sin_x[0].shape)
    
    for i in range(len(sin_x)):
        # if i % 2 == 0:
        #     sign = -1
        comb_sin = comb_sin + (sign/np.pow((i+1),2)) * sin_x[i]
        # comb_sin = comb_sin + (sign/i) * sin_x[i] + sign * cos_x[i]

    axes[1].plot(x[0],comb_sin,label="combined",color="blue")


    plt.show()

```
